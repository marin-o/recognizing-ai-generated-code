```python
import os
import sys
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

from data.dataset import AIGCodeSet
```

```python
data = AIGCodeSet(cache_dir='../../data/').get_dataset(split=False)
```

```python
# visualize length of code snippets in data['code']
import matplotlib.pyplot as plt
import numpy as np
lengths = [len(code) for code in data['code']]
plt.hist(lengths, bins=50)
plt.xlabel('Length of code snippets')
plt.ylabel('Frequency')
plt.title('Distribution of Code Snippet Lengths')
plt.show()

# visualization with outliers
plt.figure(figsize=(12, 6))
plt.boxplot(lengths)
plt.xlabel('Code Snippets')
plt.ylabel('Length')
plt.title('Boxplot of Code Snippet Lengths')
plt.grid(True)
plt.show()

# visualize number of outliers
outliers = [length for length in lengths if length > 1536]
print(f"Number of outliers (length > 1536): {len(outliers)}")
# visualize the distribution of code snippet lengths
plt.figure(figsize=(12, 6))
plt.hist(lengths, bins=50, color='blue', alpha=0.7)
plt.axvline(x=1536, color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Length of code snippets')
plt.ylabel('Frequency')
plt.title('Distribution of Code Snippet Lengths with Outliers Highlighted')
plt.grid(True)
plt.show()
```
